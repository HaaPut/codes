```python
urea = np.asarray([week0[:,0],week4[:,0],week6[:,0],week8[:,0],week10[:,0]])
creatinine = np.asarray( [week0[:,1],week4[:,1],week6[:,1],week8[:,1],week10[:,1]])
leucocyte = np.asarray([week0[:,2],week4[:,2],week6[:,2],week8[:,2],week10[:,2]])*1000
platelets = np.asarray([week0[:,3],week4[:,3],week6[:,3],week8[:,3],week10[:,3]])*1000
sodium = np.asarray([week0[:,4],week4[:,4],week6[:,4],week8[:,4],week10[:,4]])
potassium = np.asarray([week0[:,5],week4[:,5],week6[:,5],week8[:,5],week10[:,5]])
print(platelets)
```

```python
meanUrea = np.mean(urea, axis=1)
meanCreatinine = np.mean(creatinine, axis=1)
meanLeucocyte = np.mean(leucocyte, axis=1)
meanPlatelets = np.mean(platelets, axis=1)
meanSodium = np.mean(sodium, axis=1)
meanPotassium = np.mean(potassium, axis=1)
print(meanUrea)
```

```python
varUrea = np.var(urea);
varCreatinine = np.var(creatinine);
varLeucocyte = np.var(leucocyte);
varPlatelets = np.var(platelets);
varSodium = np.var(sodium);
varPotassium = np.var(potassium);
```

```python
figure1 = plt.figure(figsize=(20,15));
xlabels = ['Control','Week 4','Week 6', 'Week 8', 'Week 10']
xticks = range(5)
def subplot(figure,means,variances,title,index,color='#aed6f1'):
    ax = figure1.add_subplot(3,2,index)
    ax.plot(xticks,means);
    #ax.fill_between(xticks,means-variances,means+variances,color= '#aed6f1');
    ax.set_title(title)
    ax.set_xticks(xticks)
    ax.set_xticklabels(xlabels)


    
subplot(figure1,meanUrea,varUrea,'Mean Concentration of Urea',1)
subplot(figure1,meanCreatinine,varCreatinine,'Mean Concentration of Creatinine',2)
subplot(figure1,meanLeucocyte,varLeucocyte,'Mean Concentration of Leucocytes',3)
subplot(figure1,meanPlatelets,varPlatelets,'Mean Concentration of Platelets',4)
subplot(figure1,meanSodium,varSodium,'Mean Concentration of Serum Sodium',5)
subplot(figure1,meanPotassium,varPotassium,'Mean Concentration of Serum Potassium',6)

plt.show()
```
