```python
# collect together different values in a single array
urea = np.asarray([week0[:,0],week4[:,0],week6[:,0],week8[:,0],week10[:,0]])
creatinine = np.asarray( [week0[:,1],week4[:,1],week6[:,1],week8[:,1],week10[:,1]])
leucocyte = np.asarray([week0[:,2],week4[:,2],week6[:,2],week8[:,2],week10[:,2]])*1000
platelets = np.asarray([week0[:,3],week4[:,3],week6[:,3],week8[:,3],week10[:,3]])*1000
sodium = np.asarray([week0[:,4],week4[:,4],week6[:,4],week8[:,4],week10[:,4]])
potassium = np.asarray([week0[:,5],week4[:,5],week6[:,5],week8[:,5],week10[:,5]])
#print(platelets)
```

```python
# Compute mean concentration over different rats

meanUrea = np.mean(urea, axis=1)
meanCreatinine = np.mean(creatinine, axis=1)
meanLeucocyte = np.mean(leucocyte, axis=1)
meanPlatelets = np.mean(platelets, axis=1)
meanSodium = np.mean(sodium, axis=1)
meanPotassium = np.mean(potassium, axis=1)
#print(meanUrea)
```

```python
# Compute Variance over different
varUrea = np.var(urea, axis=1);
varCreatinine = np.var(creatinine, axis=1);
varLeucocyte = np.var(leucocyte, axis=1);
varPlatelets = np.var(platelets, axis=1);
varSodium = np.var(sodium, axis=1);
varPotassium = np.var(potassium, axis=1);
print(varUrea)
```

```python
figure1 = plt.figure(figsize=(20,15));
xlabels = ['Control','Week 4','Week 6', 'Week 8', 'Week 10']
xticks = range(5)
def subplot(figure,means,variances,title,index,color='#aed6f1'):
    ax = figure1.add_subplot(3,2,index)
    ax.plot(xticks,means);
    #ax.fill_between(xticks,means-variances,means+variances,color= '#aed6f1');
    ax.set_title(title)
    ax.set_xticks(xticks)
    ax.set_xticklabels(xlabels)


    
subplot(figure1,meanUrea,varUrea,'Mean Concentration of Urea',1)
subplot(figure1,meanCreatinine,varCreatinine,'Mean Concentration of Creatinine',2)
subplot(figure1,meanLeucocyte,varLeucocyte,'Mean Concentration of Leucocytes',3)
subplot(figure1,meanPlatelets,varPlatelets,'Mean Concentration of Platelets',4)
subplot(figure1,meanSodium,varSodium,'Mean Concentration of Serum Sodium',5)
subplot(figure1,meanPotassium,varPotassium,'Mean Concentration of Serum Potassium',6)

plt.show()
```

```python
figure2 = plt.figure(figsize=(20,15));
xlabels = ['Control','Week 4','Week 6', 'Week 8', 'Week 10']
xticks = range(5)
def subplot(figure,means,variances,title,index,color='#aed6f1'):
    ax = figure2.add_subplot(3,2,index)
    ax.errorbar(xticks,means,yerr=variances);
    ax.fill_between(xticks,means-variances,means+variances,color= '#aed6f1');
    ax.set_title(title)
    ax.set_xticks(xticks)
    ax.set_xticklabels(xlabels)


    
subplot(figure2,meanUrea,varUrea,'Mean Concentration of Urea with variance',1)
subplot(figure2,meanCreatinine,varCreatinine,'Mean Concentration of Creatinine with variance',2)
subplot(figure2,meanLeucocyte,varLeucocyte,'Mean Concentration of Leucocytes with variance',3)
subplot(figure2,meanPlatelets,varPlatelets,'Mean Concentration of Platelets with variance',4)
subplot(figure2,meanSodium,varSodium,'Mean Concentration of Serum Sodium with variance',5)
subplot(figure2,meanPotassium,varPotassium,'Mean Concentration of Serum Potassium with variance',6)

plt.show()
```

```python
sampleSize = 6
Groups = 5
pthresh = 0.05
#alternative
#standardDevUrea = np.sqrt(varUrea)
#netMeanUrea = np.mean(meanUrea)
#f_value = (np.sum(sampleSize *((meanUrea - netMeanUrea)**2)/(Groups - 1)))/(np.sum((urea - meanUrea[:,np.newaxis])**2)/((sampleSize*Groups) - Groups))
def f_test(data):
 return stats.f_oneway(data[0,:],data[1,:],data[2,:],data[3,:],data[4,:])

fUrea, pUrea = f_test(urea)
fCreatinine, pCreatinine = f_test(creatinine)
fLeucocyte, pLeucocyte = f_test(leucocyte)
fPlatelets, pPlatelets = f_test(platelets)
fSodium, pSodium = f_test(sodium)
fPotassium, pPotassium = f_test(potassium)

xticks = range(6)
xticklabels = ['Urea','Creatinine','Leucocyte','Platelets','Sodium','Potassium']
figure3 = plt.figure(figsize=(8,7))
ax = figure3.add_subplot(1,1,1);
ax.plot(xticks,[pUrea,pCreatinine,pLeucocyte,pPlatelets,pSodium,pPotassium],
        markersize=24,
        marker='o',
        linestyle='.',
        markerfacecolor='#aed6f1');
ax.plot(xticks,[pUrea,pCreatinine,pLeucocyte,pPlatelets,pSodium,pPotassium],'r.');
ax.plot(xticks,[pthresh]*len(xticks),'b--');
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)
ax.annotate('Threshold p value', xy=(2, pthresh), xytext=(2.5,pthresh+0.02),
            arrowprops=dict(facecolor='black', shrink=0.05,width=0.01))
ax.set_xlabel("Measured parameters")
ax.set_ylabel("p value ->")
plt.show()
```
